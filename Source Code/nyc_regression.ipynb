```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv("NYCdata.csv")
data.head()
```

```python
data.columns
```

```python
from sklearn.preprocessing import LabelEncoder
lb_make = LabelEncoder()
data["County_code"] = lb_make.fit_transform(data["County"])
data[["County", "County_code"]].head()
```

```python
data["Borough_code"] = lb_make.fit_transform(data["Borough"])
data[["Borough", "Borough_code"]].head()
```

```python
from sklearn.preprocessing import StandardScaler
std_scaler = StandardScaler()
x=data[['CensusTract', 'TotalPop', 'Men', 'Women',
       'Hispanic', 'White', 'Black', 'Native', 'Asian', 'Citizen', 'Income',
       'IncomePerCap', 'ChildPoverty', 'Professional', 'Service',
       'Office', 'Construction', 'Production', 'Drive', 'Carpool', 'Transit',
       'Walk', 'OtherTransp', 'WorkAtHome', 'Employed', 'PrivateWork',
       'PublicWork', 'SelfEmployed', 'FamilyWork', 'Unemployment',
       'NoIns18ab','Borough_code','County_code']]         
new_x=std_scaler.fit_transform(x)
trans_x=pd.DataFrame(new_x,columns=['CensusTract', 'TotalPop', 'Men', 'Women',
       'Hispanic', 'White', 'Black', 'Native', 'Asian', 'Citizen', 'Income',
       'IncomePerCap', 'ChildPoverty', 'Professional', 'Service',
       'Office', 'Construction', 'Production', 'Drive', 'Carpool', 'Transit',
       'Walk', 'OtherTransp', 'WorkAtHome', 'Employed', 'PrivateWork',
       'PublicWork', 'SelfEmployed', 'FamilyWork', 'Unemployment',
       'NoIns18ab','Borough_code','County_code'])
y=data['Poverty']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(new_x,y, test_size=0.2, random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,r2_score
mlr = LinearRegression()
mlr.fit(x_train,y_train)
print(mlr.intercept_)
print(mlr.coef_)
```

```python
from math import sqrt
y_pred = mlr.predict(x_test)
print("RMSE: " + str(round(sqrt(mean_squared_error(y_test,y_pred)),2)))
print("R_squared: " + str(round(r2_score(y_test,y_pred),2)))
```
